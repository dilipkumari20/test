speed test

```python
import numpy  as np
import time
size = 100000000

python_list = list(range(size))
start =time.time()
for i , v in enumerate(python_list):
    python_list[i] = v+i
list_time =time.time() - start

numpy_array = np.arange(size)
start =time.time()
result_array =numpy_array + 1
numpy_time =time.time() - start

print(f"Python list: {list_time:.4f} seconds")
print(f"NumPy array: {numpy_time:.4f} seconds")
print(f"NumPy is {list_time / numpy_time:.0f}x faster! 🚀")
```

```python
import numpy  as np
import time
size = 100

python_list = list(range(size))
start =time.time()
for i , v in enumerate(python_list):
    python_list[i] = v+i
list_time =time.time() - start

numpy_array = np.arange(size)
start =time.time()
result_array =numpy_array + 1
numpy_time =time.time() - start

print(f"Python list: {list_time:.4f} seconds")
print(f"NumPy array: {numpy_time:.4f} seconds")
print(f"NumPy is {list_time / numpy_time:.0f}x faster! 🚀")
```
